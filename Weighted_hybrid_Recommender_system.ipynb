```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
movie_data=pd.read_csv("/content/tmdb_5000_movies.csv")
credit_data=pd.read_csv("/content/tmdb_5000_credits.csv")
```

```python
movie_data.head()
```

```python
credit_data.head()
```

```python
print(credit_data.shape)
print(movie_data.shape)
```

```python
credit_data=credit_data.rename(index=str,columns={'movie_id':'id'})
dataset=movie_data.merge(credit_data,on='id')
dataset.head(3)
```

```python
dataset.columns
```

```python
dataset_cleaned=dataset.drop(columns=['homepage','title_x','title_y','production_countries'])
```

```python
dataset_cleaned.isnull().sum()
```

```python
dataset_cleaned.info()
```

```python
v=dataset_cleaned['vote_count']
R=dataset_cleaned['vote_average']
C=dataset_cleaned['vote_average'].mean()
m=dataset_cleaned['vote_count'].quantile(0.70)
```

```python
dataset_cleaned['weighted_average']=((R*v)+(C*m))/(v+m)
```

```python
dataset_cleaned.head(3)
```

```python
sorted_dataset=dataset_cleaned.sort_values('weighted_average',ascending=False)
movie_sorted=sorted_dataset[['original_title','vote_count','vote_average','weighted_average','popularity']]
movie_sorted.head(5)
```

### Recommender system based on 50% weighted_average and 50% popularity

```python
mm=MinMaxScaler()
movie_scaled=mm.fit_transform(dataset_cleaned[['weighted_average','popularity']])
movie_scaled_df=pd.DataFrame(movie_scaled,columns=['normalized_weighted_average','normalized_popularity'])
movie_scaled_df.head(5)
```

```python
df_combined = pd.concat([dataset_cleaned.reset_index(drop=True), movie_scaled_df.reset_index(drop=True)], axis=1)
df_combined.head(6)
```

```python
df_combined['score']=df_combined['normalized_weighted_average']*0.50 +df_combined['normalized_popularity']*0.50
dataset=df_combined.sort_values('score',ascending=False)
dataset=dataset[['original_title','vote_count','vote_average','normalized_popularity','normalized_weighted_average','score']]
dataset.head(5)
```
